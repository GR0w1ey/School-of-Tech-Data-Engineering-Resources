## Part 2

```python
import csv

with open('titanic-data.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    titanic_data = {'age': [], 'sex': [], 'pclass': [], 'survived': []}
    for row in reader:
        if row['age'] != '?':
            titanic_data['age'].append(float(row['age']))
        titanic_data['sex'].append(row['sex'])
        titanic_data['pclass'].append(int(row['pclass']))
        titanic_data['survived'].append(row['survived'])
```

## Part 3

```python
import matplotlib.pyplot as plt

plt.hist(titanic_data['age'], bins=20)
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## Part 4

```python
from collections import Counter

gender_counts = Counter(titanic_data['sex'])

plt.bar(gender_counts.keys(), gender_counts.values())
plt.title('Gender Distribution of Titanic Passengers')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.show()
```

## Part 5

```python
class_counts = Counter(titanic_data['pclass'])

plt.pie(class_counts.values(), labels=class_counts.keys(), autopct='%1.1f%%')
plt.title('Class Distribution of Titanic Passengers')
plt.show()
```

## Part 6 (Extension)

```python
from collections import Counter

# Calculate the number of survivors and total passengers in each class
total_passengers_by_class = Counter(titanic_data['pclass'])
survivors_by_class = Counter([pclass for pclass, survived in zip(titanic_data['pclass'], titanic_data['survived']) if survived == '1'])

# Calculate the survival rates for each class
survival_rates = {
    pclass: survivors_by_class[pclass] / total_passengers_by_class[pclass]
    for pclass in total_passengers_by_class
}

# Create the bar chart
x = ["1st Class", "2nd Class", "3rd Class"]
y = [survival_rates[pclass] for pclass in sorted(survival_rates)]
plt.figure(figsize=(10, 5))
plt.bar(x, y)
plt.title("Survival Rates by Class")
plt.xlabel("Class")
plt.ylabel("Survival Rate")
plt.ylim(0, 1)
plt.show()
```
