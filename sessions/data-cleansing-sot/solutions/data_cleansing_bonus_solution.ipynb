```python
import pandas as pd
import numpy as np
from math import isnan
```

```python
df = pd.read_csv('british-library-books.csv')

to_drop = [
    'Edition Statement',
    'Corporate Author',
    'Corporate Contributors',
    'Former owner',
    'Engraver',
    'Contributors',
    'Issuance type',
    'Shelfmarks',
    'Flickr URL'
]

df.drop(columns=to_drop, inplace=True)
df = df.set_index('Identifier')
```

```python
col = 'Date of Publication'

for i in range(len(df[col])):
    current = df[col].iloc[i]
    
    # If type of value is float, change to NaN
    if (type(current).__name__ == 'float'):
        df[col].iloc[i] = np.nan
        
    # If value contains ? or ., change to NaN
    elif '?' in current or '.' in current:
        df[col].iloc[i] = np.nan
    
    # If value contains one square bracket but not the other, change to NaN
    elif '[' in current and not ']' in current:
        df[col].iloc[i] = np.nan
    elif ']' in current and not '[' in current:
        df[col].iloc[i] = np.nan
    
    # If value starts with a [, take the integers from index 1-5 in the value (i.e. 1880)
    elif current.startswith('['):
        df[col].iloc[i] = current[1:5]
    
    
    # if date looks like 1879 [1878], remove extra date(s) and bracket(s)
    elif "[" in current or "-" in current or "," in current:
        df[col].iloc[i] = current[0:4]
```

```python
df[col] = pd.to_numeric(df[col])
df[col].dtype
```
